```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Project.csv")
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
# Total Sales Over Time
plt.figure(figsize=(10, 6))
sns.lineplot(x='transaction_date', y='Total_Bill', data=df)
plt.title('Total Sales Over Time')
plt.xlabel('Transaction Date')
plt.ylabel('Total Bill')
```

```python
# Sales Distribution by Product Category
plt.figure(figsize=(15, 6))
sns.barplot(x='product_category', y='Total_Bill', data=df)
plt.title('Sales Distribution by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Bill')
```

```python
# Hourly Sales Trend
plt.figure(figsize=(10, 6))
sns.lineplot(x='Hour', y='Total_Bill', data=df)
plt.title('Hourly Sales Trend')
plt.xlabel('Hour')
plt.ylabel('Total Bill')
```

```python
#Monthly Sales Trend
plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='Total_Bill', data=df)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Bill')
```

```python
#Product Type Contribution to Total Sales
plt.figure(figsize=(10, 6))
df.groupby('product_type')['Total_Bill'].sum().plot(kind='bar')
plt.title('Product Type Contribution to Total Sales')
plt.xlabel('Product Type')
plt.ylabel('Total Bill')
```

```python
#Store-wise Sales Comparison
plt.figure(figsize=(10, 6))
sns.barplot(x='store_location', y='Total_Bill', data=df)
plt.title('Store-wise Sales Comparison')
plt.xlabel('Store Location')
plt.ylabel('Total Bill')
```

```python
# Day-wise Sales Variation
plt.figure(figsize=(10, 6))
sns.barplot(x='Day Name', y='Total_Bill', data=df)
plt.title('Day-wise Sales Variation')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
```
